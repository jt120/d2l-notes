# 分类和回归

上一节我们学习的回归问题，预测一个数值，目标也是一个数值。

还有一类问题，就是分类，一个动物照片，让你判断它是哪种动物。

做分类，需要2个元素
1. n个类别的输出
2. 归一化，n个中最大的表示对应的类别

所以就是一个神经网络加激活函数，这个激活函数就是softmax

$\hat{y}1 = \frac{ \exp(o_1)}{\sum_{i=1}^3 \exp(o_i)},\quad \hat{y}2 = \frac{ \exp(o_2)}{\sum_{i=1}^3 \exp(o_i)},\quad \hat{y}3 = \frac{ \exp(o_3)}{\sum_{i=1}^3 \exp(o_i)}.$

softmax的输出满足

1. $\hat{y}_1 + \hat{y}_2 + \hat{y}_3 = 1$
2. $0 \leq \hat{y}_1, \hat{y}_2, \hat{y}_3 \leq 1$

所以可以把softmax的输出理解成对应类别的概率

此处说明：神经网络的原始输出是logit

## 交叉熵

回归用平方损失函数，分类用交叉熵损失函数

本质就是如果类别一致，loss就越小

$H\left(\boldsymbol y^{(i)}, \boldsymbol {\hat y}^{(i)}\right ) = -\sum_{j=1}^q y_j^{(i)} \log \hat y_j^{(i)},$

n个样本，再除以样本数

$\ell(\boldsymbol{\Theta}) = \frac{1}{n} \sum_{i=1}^n H\left(\boldsymbol y^{(i)}, \boldsymbol {\hat y}^{(i)}\right ),$
